Telling Stories with Pre-Snap Motions
=====================================

The theme for this year's Big Data Bowl is predicting post-snap results by analyizing pre-snap behavior. One of the most clear ways to visualize movement is plotting **speed** over **time** between the line set and the ball snap. Mapping the velocity shows when players are making large or small motions across the field.

This notebook breaks down four plays: the first two show pre-snap motion that exploit weaknesses in the defense, leading to offensive success and a significant number of yards gained. The second two plays show pre-snap defensive motion that effectively cover the offense and lead to an incomplete pass and an interception.

# Game: 2022091106, Play: 442

The Dolphins under Mike McDaniels are a great example of a team using motion to confuse defenses and gain an advantage before the snap. Midway through the 1st Quarter, the Dolphins are 1st and 10 at the Patriots' 45 yard line. Tyreek Hill lines up wide left of the Quarterback (Tua Tagovailoa), and just after frame 10, he starts to move quickly to the other side of the field. Hill successfully draws Jonathan Jones into man coverage and a favorable match-up. With some quick deceleration and turns, Hill is able to outmaneuver Jones, get open and convert to a successful catch.

There is information lacking in this particlar graph, though. At the time of the snap, Tyreek Hill and Jaylen Waddle are essentially "stacked" as receivers. As Jones is covering Hill, Jalen Mills moves to cover Waddle, but it appears to be an overall unfavorable situation for both defensive players. 

Running back Chase Edmonds is also open as Tagovailoa makes his pass, but Hill is already well past the line of scrimmage, making his catch more valuable than if Tagovailoa makes a short (but safe) pass to Edmonds.

The Kansas City Chiefs run a similar play in their game against the Arizona Cardinals on September 11th (gameid: 2022091110, playid: 514), where TE Travis Kelce moves across the field from left to right before the snap. However, instead of Kelce being the receiver, he moves the defense back, creating space for Clyde Edwards-Helaire to make a reception for 25 total yards.

```python
plot_speed_over_time(2022091106,442)
```

# Game: 2022091103, Play: 1126

Early in the second quarter, on 1st and 10, Zach Gentry (TE) moves from left of the QB (Mitch Trubisky) to the right. We don't see any significant movement from the defense over the corresponding frames. The defense appears to be confident that their existing formation doesn't need to shift to accomodate Gentry's new position. 

At frame 106, George Pickens starts moving from right of the QB to his left. Here, we can clearly see the defense react to George Pickens' movement. First Chidobe Awuzie follows in partial man coverage, but then drops off around frame 120. At the same time, we see Vonn Bell, Germaine Pratt and Logan Wilson shifting in zone coverage.

George Pickens' motion seems to successfully draw the attention of the defense, and they move to cover him, leaving Zach Gentry with reduced coverage. When Gentry moves past the defender, he's wide open to catch Trubisky's pass and gains 32 yards on the catch.

```python
plot_speed_over_time(2022091103,1126)
```

# Game: 2022091101, Play: 339

By the graph, this play looks somewhat similar to the Tyreek Hill motion play above, but further analysis reveals some differences.

Early in the first quarter, the play is 4th and 3 as the Cleveland Browns are on the 36 yard line of the Carolina Panthers. Donovan Peoples-Jones goes into motion, moving from left to right and bringing defender Donte Jackson along with him. However, where Tyreek Hill stays in motion at the time of the snap, Peoples-Jones comes to a full stop with Jackson in front of him. Peoples-Jones pulls Jackson across two where two other defenders are already lined up. Yetur Gross-Moss and Jeremy Chinn are double covering David Njoku on the line of scrimmage. 

Carolina sticks to man coverage to make sure that Cleveland doesn't get the first down. Once the snap happens, all of the defenders stay on top of their assigned receivers. Jackson is able to break up the pass, and the Panthers take possession of the ball in their own territory.

# Conclusion

Mapping speed over time gives some deep insights into motion patterns before the snap, but clearly there are still some gaps. The next step would be to look at the Y values, to see how defenders and receivers are lining up, to get a better sense of who is covering who. There are also opportunities in narrowing down what makes a "good outcome" for well designed plays. If a player can gain a large number of yardage after the catch (like Gentry above), that looks great. But that also has a lot to do with Gentry's ability to avoid getting tackled. That ability isn't really a factor when examining the contributions of pre-snap movement.
